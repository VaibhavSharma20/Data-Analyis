```python
import pandas as pd
import matplotlib as plt
from scipy.stats import pearsonr

data=pd.read_csv("Iris.csv")
data.shape
corr, _ =pearsonr(data.SepalLengthCm,data.SepalWidthCm)
print("Pearson Correlation is: %f" %corr)
```

```python
import pandas as pd
import matplotlib as plt
from scipy.stats import spearmanr

data=pd.read_csv("Iris.csv")
data.shape
corr, _ =spearmanr(data.SepalLengthCm,data.SepalWidthCm)
print("Spearman Correlation is: %f" %corr)
```
