```python
import numpy as np
import pandas as pd
from pandas import plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Mall.csv')
data.head()
```

```python
data.rename(columns={'Genre':'Gender','Annual Income (k$)':'AnnualIncome','Spending Score (1-100)':'SpendingScore'},inplace=True)
data.head()
```

```python
genders = data.Gender.value_counts()
plt.figure(figsize=(10,4))
plt.title('Gender')
sns.barplot(x=genders.index, y=genders.values)
plt.show()
```

There are more females
112 female and 88 male

```python
data['Age'].value_counts().plot.bar(figsize = (18, 4),grid=False,legend=True,fontsize=15,color='red')
plt.title('Distribution of age', fontsize = 17)
plt.show()
```

People at Age 32 are the Frequent Visitors in the Mall.
People of older age,above 50s groups are lesser frequent in the Malls 
Ages from 19 and 31 are very much frequent

```python
data['AnnualIncome'].value_counts().plot.bar(figsize = (18, 4),grid=False,legend=True,fontsize=15,color='green')
plt.title('Distribution of annual income', fontsize = 17)
plt.show()
```

varity of income is visible (15k to 137k)
People above 100k annual income are 12
People below 25k anual income are 22
majority of population has annual income between 25k and 100k

```python
data['SpendingScore'].value_counts().plot.bar(figsize = (18, 4),grid=False,legend=True,fontsize=15,color='blue')
plt.title('Distribution of spending score', fontsize = 17)
plt.show()
```

Mall caters to people having spending score ranging from 1-100.
majority of the customers have spending score between 35 and 75

```python
ax = sns.boxplot(data=data[['Age','AnnualIncome','SpendingScore']], orient="h", palette=sns.color_palette("hls", 5))
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
sns.heatmap(data.corr(), cmap = 'cool', annot = True,square=False)
plt.title('Heatmap for the Data', fontsize = 17)
plt.show()
```

```python
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(data=data,hue='Gender',palette='Paired',diag_kind='hist',markers='s')
plt.show()
```

```python
x = data['AnnualIncome']
y = data['Age']
z = data['SpendingScore']

sns.lineplot(x, y, color = 'blue')
sns.lineplot(x, z, color = 'red')
plt.title('Annual Income vs Age and Spending Score', fontsize = 20)
plt.show()
```
